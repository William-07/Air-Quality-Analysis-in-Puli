```python
import pandas as pd
from datetime import datetime
df_Hourly_Air_Data = pd.read_csv('/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Concated_Data/Hourly_Air_Data.csv')
```

```python
df_Hourly_Air_Data
```

```python
#Check null values of each attribute
print(df_Hourly_Air_Data.isnull().sum())
```

```python
#drop attributes which are nominal or too have number of missing value > 5000
df_Hourly_Air_Data = df_Hourly_Air_Data.drop(columns = ['County','SiteName','Pollutant','Status','SO2_AVG','Longitude','Latitude','SiteId','Unit'])
#df_Puli_data
```

```python
#remove the cells which are 'ND' in column PM2.5
df_Hourly_Air_Data = df_Hourly_Air_Data[df_Hourly_Air_Data['PM2.5'] != 'ND']

#Change the data type of PM2.5 from string to float
df_Hourly_Air_Data = df_Hourly_Air_Data.astype({'PM2.5':'float'})
```

```python
#fill missing value with interpolation
df_Hourly_Air_Data = df_Hourly_Air_Data.interpolate() 
```

```python
#reset the index
df_Hourly_Air_Data = df_Hourly_Air_Data.reset_index(drop=True)
```

```python
df_Hourly_Air_Data
```

```python
#change the datatype from string to datetime
df_Hourly_Air_Data = df_Hourly_Air_Data.copy()
for i in range(len(df_Hourly_Air_Data['DataCreationDate'])):
    df_Hourly_Air_Data['DataCreationDate'][i] = datetime.strptime(df_Hourly_Air_Data['DataCreationDate'][i],'%Y-%m-%d %H:%M')
```

```python
df_Hourly_Air_Data
```

```python
# chage data in daily timeframe
date = []
df_Daily_Air_Data = df_Hourly_Air_Data.copy()
for d in df_Daily_Air_Data['DataCreationDate']:
    date.append(d.strftime("%Y-%m-%d"))
df_Daily_Air_Data['Date'] = date
#Calculate the daily average air quality
df_Daily_Air_Data = df_Daily_Air_Data.groupby(df_Daily_Air_Data.Date).mean().reset_index()
```

```python
df_Daily_Air_Data
```

```python
#Classify the hourly air quality data into four season(1.early spring(Feb~April) 2.late spring & summer(May~Sep) 3.Autumn(Oct~Nov) 4.Winter(Dec~Jan))
for i in range(len(df_Hourly_Air_Data['DataCreationDate'])):
    
#     print(df['DataCreationDate'][i].month)
    
    if df_Hourly_Air_Data['DataCreationDate'][i].month == 2 or df_Hourly_Air_Data['DataCreationDate'][i].month == 3 or df_Hourly_Air_Data['DataCreationDate'][i].month == 4:
        df_Hourly_Air_Data['DataCreationDate'][i] = "Early Spring"
        
    elif df_Hourly_Air_Data['DataCreationDate'][i].month == 5 or df_Hourly_Air_Data['DataCreationDate'][i].month == 6 or df_Hourly_Air_Data['DataCreationDate'][i].month == 7 or df_Hourly_Air_Data['DataCreationDate'][i].month == 8 or df_Hourly_Air_Data['DataCreationDate'][i].month == 9:
        df_Hourly_Air_Data['DataCreationDate'][i] = "Late Spring & Summer"
    
    elif df_Hourly_Air_Data['DataCreationDate'][i].month == 10 or df_Hourly_Air_Data['DataCreationDate'][i].month == 11:
        df_Hourly_Air_Data['DataCreationDate'][i] = "Autumn"
    
    elif df_Hourly_Air_Data['DataCreationDate'][i].month == 12 or df_Hourly_Air_Data['DataCreationDate'][i].month == 1:
        df_Hourly_Air_Data['DataCreationDate'][i] = "Winter"

df_Hourly_Air_Data = df_Hourly_Air_Data.rename(columns={"DataCreationDate":"Season"})
df_Hourly_Air_Data   

```

```python
#Export Hourly_Air_Data.csv
df_Hourly_Air_Data.to_csv("/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Preprocessing_Data/Hourly_Air_Data.csv", index = False)
```

```python
#Classify the daily air quality data into four season(1.early spring(Feb~April) 2.late spring & summer(May~Sep) 3.Autumn(Oct~Nov) 4.Winter(Dec~Jan))

for i in range(len(df_Daily_Air_Data['Date'])):
    df_Daily_Air_Data['Date'][i] = datetime.strptime(df_Daily_Air_Data['Date'][i],'%Y-%m-%d')

for i in range(len(df_Daily_Air_Data['Date'])):
    
#     print(df['DataCreationDate'][i].month)
    
    if df_Daily_Air_Data['Date'][i].month == 2 or df_Daily_Air_Data['Date'][i].month == 3 or df_Daily_Air_Data['Date'][i].month == 4:
        df_Daily_Air_Data['Date'][i] = "early spring"
        
    elif df_Daily_Air_Data['Date'][i].month == 5 or df_Daily_Air_Data['Date'][i].month == 6 or df_Daily_Air_Data['Date'][i].month == 7 or df_Daily_Air_Data['Date'][i].month == 8 or df_Daily_Air_Data['Date'][i].month == 9:
        df_Daily_Air_Data['Date'][i] = "late spring & summer"
    
    elif df_Daily_Air_Data['Date'][i].month == 10 or df_Daily_Air_Data['Date'][i].month == 11:
        df_Daily_Air_Data['Date'][i] = "Autumn"
    
    elif df_Daily_Air_Data['Date'][i].month == 12 or df_Daily_Air_Data['Date'][i].month == 1:
        df_Daily_Air_Data['Date'][i] = "Winter"
        
df_Daily_Air_Data = df_Daily_Air_Data.rename(columns={"Date":"Season"})

df_Daily_Air_Data 

```

```python
#Export Daily_Air_Data.csv
df_Daily_Air_Data.to_csv("/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Preprocessing_Data/Daily_Air_Data.csv",index = False)
```

```python
#Weekly Air Quality Data
#Get weely data's season:

#1. Transfer the nominal season name to integer
Season_code = []
df_Weekly_Air_Data = df_Daily_Air_Data.copy()
for s in df_Daily_Air_Data['Season']:
    if s =='Winter':
        Season_code.append(1)
        
    elif s =='early spring':
        Season_code.append(2)
        
    elif s =='late spring & summer':
        Season_code.append(3)
        
    elif s =='Autumn':
        Season_code.append(4)
df_Weekly_Air_Data['Season_code'] = Season_code

#2. group data into weekly (7 days) format
Days = 7
Season_code = []
#sum the seasonal code
Season_code = df_Weekly_Air_Data.groupby(df_Weekly_Air_Data.index // Days).sum()['Season_code']
df_Weekly_Air_Data = df_Weekly_Air_Data.groupby(df_Weekly_Air_Data.index // Days).mean()
df_Weekly_Air_Data['Season_code'] = Season_code

#3. assign season to each week
Season = []
season = " "
for sc in df_Weekly_Air_Data['Season_code']:
    if sc/7 == 1:
        season = 'Winter'
    elif sc/7 == 2:
        season = 'early spring'
        
    elif sc/7 ==3:
        season = 'late spring & summer'
        
    elif sc/7 ==4:
        season = 'Autumn'
    #if %7 != 0, the season same with the last one    
    else:
        season = season
    Season.append(season)
    
df_Weekly_Air_Data['Season'] = Season
df_Weekly_Air_Data = df_Weekly_Air_Data.drop(columns = ['Season_code'])
df_Weekly_Air_Data
        


        
```

```python
#Export Weekly_Air_Data.csv
df_Weekly_Air_Data.to_csv("/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Preprocessing_Data/Weekly_Air_Data.csv", index = False)
```

```python
#Monthly Air Quality Data(30 days interval)
#Get Monthly data's season:

#1. Transfer the nominal season name to integer
Season_code = []
df_Monthly_Air_Data = df_Daily_Air_Data.copy()
for s in df_Daily_Air_Data['Season']:
    if s =='Winter':
        Season_code.append(1)
        
    elif s =='early spring':
        Season_code.append(2)
        
    elif s =='late spring & summer':
        Season_code.append(3)
        
    elif s =='Autumn':
        Season_code.append(4)
df_Monthly_Air_Data['Season_code'] = Season_code

#2. group data into monthly (30 days) format
Days = 30
Season_code = []
Season_code = df_Monthly_Air_Data.groupby(df_Monthly_Air_Data.index // Days).sum()['Season_code']
df_Monthly_Air_Data = df_Monthly_Air_Data.groupby(df_Monthly_Air_Data.index // Days).mean()
df_Monthly_Air_Data['Season_code'] = Season_code

#3. assign season to each Month
Season = []
season = " "
for sc in df_Monthly_Air_Data['Season_code']:
    if sc/30 == 1:
        season = 'Winter'
    elif sc/30 == 2:
        season = 'early spring'
        
    elif sc/30 ==3:
        season = 'late spring & summer'
        
    elif sc/30 ==4:
        season = 'Autumn'
    #if %7 != 0, the season same with the last one    
    else:
        season = season
    Season.append(season)
df_Monthly_Air_Data['Season'] = Season
df_Monthly_Air_Data = df_Monthly_Air_Data.drop(columns = ['Season_code'])
df_Monthly_Air_Data
```

```python
#Export Monthly_Air_Quality_Data.csv
df_Monthly_Air_Data.to_csv("/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Preprocessing_Data/Monthly_Air_Data.csv", index = False)
```
