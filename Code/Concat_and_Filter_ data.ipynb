```python
import pandas as pd 
import numpy as np
import os

#concate all csv file
#Get the list of all files
path = '/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/All_Data'
data_files = os.listdir(path)
#mac os has a specific 
if '.DS_Store' in data_files:
    data_files.remove('.DS_Store')
#append all the data in the df_list
df_list = []
for i in data_files:
    df_list.append(pd.read_csv(path + "/" + i))
#Turn into Pandas dataframe
Original_data = pd.concat(df_list, ignore_index=True)

#filter out the data containing "埔里" ("Puli")
mask = Original_data["SiteName"] == "埔里"
#Sort Data by time
Hourly_Air_Data = Original_data[mask].sort_values(by = "DataCreationDate")
#reset the index
Hourly_Air_Data = Hourly_Air_Data.reset_index(drop=True)
#Export Data
Hourly_Air_Data.to_csv(r'/Users/shenglienlee/Desktop/GitHub/Air-Quality-Analysis-in-Puli/Data/Concated_Data/Hourly_Air_Data.csv', index = False,encoding='utf_8_sig')
Hourly_Air_Data
```
